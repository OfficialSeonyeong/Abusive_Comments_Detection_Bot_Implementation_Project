```python
from selenium import webdriver
import time

path='C:/Users/A/TIL/07_웹크롤링/driver/chromedriver.exe'
driver = webdriver.Chrome(path)
url = 'https://everytime.kr/'
driver.get(url)



time.sleep(5)
```

```python
# 로그인 버튼 누르기
login = driver.find_element_by_css_selector('.login').find_elements_by_tag_name('a')
login[1].click()

time.sleep(5)

# 아이디 입력
id_input = driver.find_element_by_tag_name('div > form > p > input')
id_input.send_keys('xhfl001')


inputs = driver.find_element_by_tag_name('div > form').find_elements_by_tag_name('input')

# 비밀번호 입력
password_input = inputs[1]
password_input.send_keys('비밀번호입력')

time.sleep(1)

# 로그인 실행
inputs[4].submit()

time.sleep(5)
```

```python
# 게시글 들어갔다 나오기
# total_comments = []

page_list = [14,15,16]


for page in page_list :
    
    print(page, '번째 페이지 입니다.')
    url = "https://everytime.kr/370441/p/" + str(page)
    driver.get(url)
    
    time.sleep(3)
    
    posts = driver.find_elements_by_tag_name('body > #container > div.articles > article')
    n = len(posts)
    
    for idx in range(n):
        
        print(idx+1, '번째 게시물 입니다.')
        posts[idx].find_element_by_tag_name('a.article').send_keys('\n')
        time.sleep(3)

        # 댓글 가져오기
        try :
            comments = driver.find_elements_by_css_selector('.comments > .parent > .large')
            for comment in comments:
                total_comments.append(comment.text)
        except :
            print('댓글이 존재하지 않습니다')

    

        try :
            re_comments = driver.find_elements_by_css_selector('.comments > .child > .large')
            for comment in re_comments:
                total_comments.append(comment.text)
        except :
            print('대댓글이 존재하지 않습니다.')


        driver.back()

        time.sleep(3)
        posts = driver.find_elements_by_tag_name('body > #container > div.articles > article')
    
    print('현재 까지 크롤링한 댓글 수 :', len(total_comments))

        
        
        

# print(total_comments)
```

```python
import pandas as pd

everytime_comments_df = pd.DataFrame({'댓글' : total_comments})

everytime_comments_df.to_csv('./에브리타임_댓글_1113.csv', encoding='utf-8-sig')
```
